## 数据预处理
顺序：
1.删除重复数据、删除异常数据
2.处理累计数据、筛选交易时段数据、其他处理
原因：先进行数据的删除，再按需进行筛选等其他工作

数据定义：
重复数据：1.包括datetime在内的每一列数据都相同；2.两行数据间隔小于三秒，并且其他字段数据相同。满足二者其一则为重复数据
异常数据：1.买一价或卖一价为0；2.空值（nan）
累计数据：某个时间段内的累计数据，如一分钟内的成交量
交易时段数据：开盘至收盘，即9：30-11：30，13：00-15：00，不包括集合竞价阶段
